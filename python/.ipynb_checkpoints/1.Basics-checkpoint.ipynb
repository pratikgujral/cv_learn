```python
## Imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Load and display images

```python
im = cv2.imread('../input_files/watch.jpg'); ## Loads image as is.
cv2.imshow('Original image', im)

grayed_im = cv2.imread('../input_files/watch.jpg', 0) # Loads image in grayscale
cv2.imshow('Grayscaled image', grayed_im)

cv2.waitKey(0) ## Waits for a key press for the specified time(ms). If 0, it waits indefinitely.
cv2.destroyAllWindows()
```

Second argument to imread() is a flag which specifies the way image should be read.

cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
cv2.IMREAD_GRAYSCALE (or 0) : Loads image in grayscale mode
cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

```python
## Closing the image 'window' on a specific key press
im = im = cv2.imread('../input_files/watch.jpg'); ## Loads image as is.
cv2.imshow('Original image', im)
k = cv2.waitKey(0)
if k == 27 & 0xFF: ## ASCII equivalent for Esc key == 27
    cv2.destroyAllWindows()
```

```python
## Using matplotlib to display images
plt.imshow(im, interpolation = 'bicubic')

## Remove the xticks and yticks
plt.xticks([]), plt.yticks([])
```

Note that the colors in the images displayed by cv2 and matplotlib are not the same. This is because of the way colors are treated differently by the two:
1. matplotlib -> RGB
2. OpenCV -> BGR

```python
print(im.shape)
## Splitting the B,G,R components
b, g, r = cv2.split(im)

print(b.shape, g.shape,  r.shape)

im2 = cv2.merge([r,g,b])
## Now displaying using matplotlib
plt.imshow(im2, interpolation='bicubic')
plt.show()
```

```python
## Displaying size, shape of image
print(im.shape)

nrows, ncolumns, nchannels = im.shape
print('Rows: ', nrows, 'Cols: ', ncolumns, 'Channels: ', nchannels)
```

# ROI

```python
## Cropping an image
im = cv2.imread('../input_files/watch.jpg')
cv2.imshow('Original Image',im)

dial = im[100:250, 280:450, :]
cv2.imshow('dial', dial)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Video

```python
cap = cv2.VideoCapture(0) ## To read from file, pass the path to file as a string to the VideoCapture method
print(cap.isOpened())

while cap.isOpened():
    ret, frame = cap.read()
    if ret: # If a frame is returned
        cv2.imshow('frame', frame)
    k = cv2.waitKey(1) & 0xFF
    if k == 27: # Esc Key
        break

cap.release()
cv2.destroyAllWindows()
```
